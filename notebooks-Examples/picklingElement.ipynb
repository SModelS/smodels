```python
import sys
sys.path.append('../')
import networkx as nx
import matplotlib.pyplot as plt
from smodels.theory.element import Element
from smodels.experiment.defaultFinalStates import finalStates
from smodels.theory.tree import Tree, compareNodes
from smodels.particlesLoader import BSMList
from smodels.share.models.SMparticles import SMList
from smodels.theory.model import Model
from smodels.theory.auxiliaryFunctions import index_bisect
from collections import OrderedDict
from smodels.tools.physicsUnits import fb, GeV
import pickle as serializer
```

```python
elString = "[[['jet']],[['jet', 'e-'],['nu','e-']]]"
el = Element(elString,model=finalStates)
```

```python
el.drawTree()
print(el.tree.nodes)
print(el.tree.edges)
```

```python
testFile = "test_element.pcl"
with open(testFile,"wb") as f:
    serializer.dump(el,f,protocol=4)
```

```python
with open(testFile,"rb") as f:
    newEl = serializer.load(f)
```

```python
newEl.drawTree()
print(newEl.tree.nodes)
print(newEl.tree.edges)
```
