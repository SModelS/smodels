### Create element using a BSM model

```python
slhafile = '../inputFiles/slha/lightEWinos.slha'
model = Model(BSMparticles=BSMList, SMparticles=SMList)
model.updateParticles(inputFile=slhafile)
```

```python
stringEl = "(PV > su_L(1),su_L~(2)), (su_L(1) > gluino(3),e-,nu), (su_L~(2) > C1+,d), (gluino(3) > N1,u,u~)"
el = ExpSMS.from_string(stringEl, model=model)
el.draw()
print(el.nodes)
print(el.nodeIndices)
print(el.edges)
```

#### Draw tree using distinct property for labels

```python
el.draw(nodeScale=4,labelAttr='isSM')
```

```python
el.draw(nodeScale=4,labelAttr='mass',attrUnit=GeV)
```

```python
el.draw(nodeScale=5,labelAttr='node')
```

#### Convert to old bracket notation

```python
elB,finalState,intermediateState = el.treeToBrackets()
print(elB)
print(finalState)
print(intermediateState)
```

```python
stringEl = "(PV > N1,C1+(1)), (C1+(1) > N1,u,d)"
el = ExpSMS.from_string(stringEl,model=model)
el.draw()
print('Name=',el.canonName,'\nProcess=',el)
print([(n,nIndex) for n,nIndex in zip(el.nodes,el.nodeIndices)])
print(el.edges)
```

```python
elB,finalState,intermediateState = el.treeToBrackets()
print(elB)
print(finalState)
print(intermediateState)
```
