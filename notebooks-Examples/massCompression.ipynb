```python
import sys
sys.path.append('../')
import networkx as nx
from networkx import bfs_successors
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
from smodels.theory.element import Element
from smodels.particlesLoader import BSMList
from smodels.share.models.SMparticles import SMList
from smodels.theory.model import Model
from smodels.tools.physicsUnits import GeV
import itertools
```

```python
slhafile = '../inputFiles/slha/higgsino_spectrum_520_125_dm_4.slha'
model = Model(BSMparticles=BSMList, SMparticles=SMList)
model.updateParticles(inputFile=slhafile, promptWidth=1e-10*GeV)
```

```python
stringEl = "(PV > C2+(1),N3(2)), (N3(2) > C1-(3),W+), (C1-(3) > N1,e-,nu), (C2+(1) > N2(4),W+), (N2(4) > N1(5),nu,nu), (N1(5) > sta_1,ta+)"
el = Element(info=stringEl,model=model)
el.drawTree(nodeScale=3)
print(el.tree.nodes)
```

```python
el.drawTree(nodeScale=3,labelAttr='mass',attrUnit=GeV)
print(el.mass)
```

```python
newEl = el.massCompress(5*GeV)
newEl.drawTree()
print(newEl.tree.nodes)
print(newEl.mass)
```

```python
newEl = el.massCompress(3*GeV)
print(newEl)
```
